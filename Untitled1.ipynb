```python
import matplotlib as mpl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
mpl.rcParams['figure.figsize'] = (10,10)

heroes_data = pd.read_csv("../Downloads/superhero-set/heroes_information.csv")

heroes_powers = pd.read_csv("../Downloads/superhero-set/super_hero_powers.csv")


total_heroes = len(set(heroes_data['name']))

map_gender = {}

for i in range(len(heroes_data['Gender'])):
	map_gender[heroes_data['name'][i]] = heroes_data['Gender'][i] 	


total_power = heroes_powers.keys()[1:]

fig, ax = plt.subplots()

bar_width = 0.35
opacity = 0.4
error_config = {'ecolor': '0.3'}
num_of_shared_power = 0
num_of_heroes_for_power = []
for i in total_power:
	count_a_power = list(heroes_powers[i]).count(True)
	if count_a_power < 2:
		num_of_shared_power += 1
	num_of_heroes_for_power.append(count_a_power)

print("Total number of unique power - " + str(len(total_power)))
print()
print()
print("Total number of shared power - " + str(num_of_shared_power)) 
print()
print()
print("Total number of heroes - " + str(total_heroes))

rects1 = ax.bar(np.arange(len(num_of_heroes_for_power)), num_of_heroes_for_power, bar_width
                ,alpha=opacity, color='b',
                 error_kw=error_config,
                label='Heroes')

ax.set_xlabel('Powers')
ax.set_ylabel('Number of heroes')
ax.set_title('chart explaining no of heroes share a specific power')
ax.set_xticks(np.arange(len(num_of_heroes_for_power)) +  bar_width/2  )
ax.set_xticklabels(total_power[1:])
for tick in ax.xaxis.get_major_ticks():
	tick.label.set_fontsize(4) 
	# specify integer or one of preset strings, e.g.
	#tick.label.set_fontsize('x-small') 
	tick.label.set_rotation('vertical')

ax.legend()
fig.tight_layout()
plt.show()
```
